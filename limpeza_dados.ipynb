#### Limpeza e extração de dados

```python
### Importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Importando dados
df = pd.read_excel('Retornos_diarios.xlsx', index_col=0, parse_dates=True, skiprows=3)
df.head()
```

```python
def columns(df): #renomeando colunas com as ações
    df.columns = df.columns.str[59:]
columns(df)
df.replace('-', np.nan, inplace=True) #substituindo valores ausentes no dataframe
```

```python
nul_values = df.isnull().sum().sort_values() #calculando valores nulos
stocks = pd.Series(nul_values).where(lambda x: x==nul_values.min()).dropna() #selecionando ações que não possuem valores nulos
stocks.count()
```
